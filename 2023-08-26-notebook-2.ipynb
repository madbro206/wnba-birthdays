This code finds which WNBA players' teams had games on their birthday during the 2023 season.

```r
players <-read.table(file='players.csv', header=TRUE, sep=",")
schedule <-read.table(file='2023schedule.csv', header=TRUE, sep=",")
key <-read.table(file='team_key.csv', header=TRUE, sep=",")
```

```r
schedule$Date <- strptime(schedule$Date, format = "%a %b %d %Y")
schedule$Date <- format(schedule$Date, format = "%Y-%m-%d")
```

```r
library(dplyr)

new_df1 <- select(schedule, date = Date, team1 = Visitor.Neutral)
```

```r
new_df1
```

```r
new_df2 <- select(schedule, date = Date, team2 = Home.Neutral)
```

```r
new_df <- bind_rows(new_df1, new_df2) %>%
  mutate(team = ifelse(is.na(team1), team2, team1)) %>%
  select(date, team)
new_df
```

```r
joined_df <- left_join(new_df, key, by = "team")
joined_df
```

```r
joined_df$date <- strftime(joined_df$date, format = "%m-%d")
joined_df
```

```r
players$DOB <- strftime(players$DOB, format = "%m-%d")
players
```

```r
matching_rows <- merge(players, joined_df, by.x = c("DOB", "Team"), by.y = c("date", "abb"))
matching_rows
```
